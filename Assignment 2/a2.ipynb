## Problem 1

#### Find all the words with a given letter in a given position (simple)

```python
import os

wordlist = []

# used for the efficient solution
largest_word_length = 0

# Open words.txt file
with open("words.txt", "r") as file: 
    for word in file:
        # Remove blank space before/after words
        word = word.strip()
        if len(word) > largest_word_length: 
            largest_word_length = len(word)
        wordlist.append(word)
```

```python
def simple_search(words, pos, letter):
    # Indices start from 0
    pos = pos - 1
    matching_words = []
    for word in words:
        # If the position is bigger than the length of the word we don't need to check further
        if len(word) > pos and word[pos] == letter:
            matching_words.append(word)

    return matching_words
```

```python
result = simple_search(wordlist, 4, "z")
print("result =", result)
```

#### Find all the words with a given letter in a given position (efficient)

We can build an index from the given position in all the words. This is done using a dictionary (map) where the keys are the positions and the values are a map (keys = all letters in that specific position, values = list of matching words). This way we can build an index for the entire list of words once and directly grab the list of words that have the given letter in the given position.

```python
def build_index(words):
    index = {}

    # Add all possible positions as keys and empty dicts as values
    for x in range(0, largest_word_length):
        index[x] = {}

    for word in words:
        # Go over each letter of the word
        for x in range(0, len(word)):
            letter = word[x]

            letter_dict = index[x]

            if letter not in letter_dict:
                letter_dict[letter] = [word]
            else:
                letter_dict[letter].append(word)
    
    return index

def efficient_search(index, pos, letter):
    # Indices start from 0
    return index.get(pos-1).get(letter)
```

```python
index = build_index(wordlist)
result = efficient_search(index, 4, "z")
print("result =", result)
```

```python
from timeit import timeit

# Call the simple solution 50 times and time it
simple_solution = timeit(lambda: simple_search(wordlist, 4, "z"), number=250)
# Call the building index once and time it
index_building_time = timeit(lambda: build_index(wordlist), number=1)
# Call the efficient solution 50 times and time it
efficient_solution = timeit(lambda: efficient_search(index, 4, "z"), number=250) + index_building_time

print(f"Executing the simple solution 250 times takes: {simple_solution}s")
print(f"Building the index takes: {index_building_time}s")
print(f"Executing the efficient solution 250 times takes: {efficient_solution}s")
```

As seen from the output above almost all of the time spent for the efficient solution is spent for building the index and the actual searching is almost instant.

## Problem 2

- **`read_words()`** - goes through all the words and adds them to a dictionary where the keys are the sorted words, e.g. `"anagram"` becomes `"aaagmnr"` and the values are the lists of matching anagrams
- **`find_anagram_lists_by_length`** - finds all the anagram groups (at least 2 matching words) and sorts them by list length
- **`find_anagram_lists_by_length_n`** - finds all the anagram groups (at least 2 matching words) where the length of the words in the anagram groups are `n`
- **`write_anagrams_to_file`** - writes the lists returned by either `find_anagram_lists_by_length` or `find_anagram_lists_by_length_n` to a desired output file (`output.txt` by default)


```python
def read_words(words):
    dict = {}
    for word in words:
        # sorted returns a list so we need to join
        sorted_word = "".join(sorted(word))
        if sorted_word not in dict:
            dict[sorted_word] = [word]
        else:
            dict[sorted_word].append(word)

    return dict

def find_anagram_lists_by_length(word_dict):
    # Add all values in the dict (anagram groups) to a list if the length is more than 1
    anagram_lists = [anagram_group for anagram_group in word_dict.values() if len(anagram_group) > 1]
    anagram_lists.sort(key=len)
    return anagram_lists

def find_anagram_lists_by_length_n(word_dict, n):
    # Add all values in the dict (anagram groups) to a list if the length is more than 1 and words in the group are of length n
    anagram_lists = [anagram_group for anagram_group in word_dict.values() if len(anagram_group[0]) == n and len(anagram_group) > 1]
    return anagram_lists

def write_anagrams_to_file(anagram_lists, file_path):
    with open(file_path, 'w') as file:
        for anagram_group in anagram_lists:
            # Write each anagram group to a separate line
            file.write(", ".join(anagram_group) + "\n")
```

```python
output_file = "output.txt"
output_file_n = "output_n.txt"

anagram_dict = read_words(wordlist)
anagram_lists_by_length = find_anagram_lists_by_length(anagram_dict)

n = int(input("Enter the value of n: "))
anagram_lists_by_length_n = find_anagram_lists_by_length_n(anagram_dict, n)

write_anagrams_to_file(anagram_lists_by_length, output_file)
write_anagrams_to_file(anagram_lists_by_length_n, output_file_n)
```

#### Which data structure did you use and why?

I used a dictionary to be able to build and access the list of anagrams easily and swiftly as lookups are constant complexity. I also used lists to store the "real" words which had the same letter structure which enables easy additions unlike tuples which are immutable.
